```python
class node:

  def __init__(self, d):
    self.data = d
    self.next = None

class clinkedlist:

  def __init__(self):
    self.head = None

  def display(self):
    t = self.head
    if t is None:
      print('empty')
      return
    while t.next != self.head:
      print(t.data, end='-->')
      t = t.next
    print(t.data)

  def addfirst(self, data):
    n = node(data)
    if self.head == None:
      self.head = n
      n.next = n
      return
    else:
      n.next = self.head
      t = self.head
      while t.next != self.head:
        t = t.next
      t.next = n
      self.head = n

  def addafter(self, m, data):
    if self.head == None :
      return -1
    else:
      n = node(data)
      t = self.head
      while t.data != m:
        t = t.data
      n.next = t.next
      t.next = n

  def delete(self,m):
    if self.head == None:
      print('empty')
      return
    elif self.head.next == self.head:
      if self.head.data == m:
        self.head == None
        return
      return -1
    else:
      t = self.head
      while t.data != m :
        p = t
        t = t.next
        if t == self.head:
          return -1
      p.next = t.next
      t.next = None

circular_linked_list  = clinkedlist()

circular_linked_list.addfirst(8)
circular_linked_list.addfirst(10)
circular_linked_list.addfirst(4)
circular_linked_list.addfirst(6)
circular_linked_list.addfirst(0)

circular_linked_list.display()

circular_linked_list.delete(10)

circular_linked_list.display()
```

```python
class dnode:

  def __init__(self,data):
    self.data = data
    self.next = None
    self.prev = None

class cdlinkedlist:

  def __init__(self):
    self.head = None

  def display(self):
    t = self.head
    while t.next != self.head:
      print(t.data, end='<-->')
      t = t.next
    print(t.data)

  def addfirst(self, data):
    n = node(data)
    if self.head == None:
      self.head = n
      n.next = n
      n.prev = n
      return
    else:
      n.next = self.head
      self.head.prev = n
      t = self.head
      while t.next != self.head:
        t = t.next
      t.next = n
      n.prev = t
      self.head = n

  def addafter(self, m, data):
    if self.head == None :
      return -1
    else:
      n = node(data)
      t = self.head
      while t.data != m:
        t = t.data
      n.next = t.next
      t.next.prev = n
      t.next = n
      n.prev = t

  def delete(self,m):
    if self.head == None:
      print('empty')
      return
    elif self.head.next == self.head and self.head.prev == self.head:
      if self.head.data == m:
        self.head == None
        return
      return -1
    else:
      t = self.head
      while t.data != m :
        t = t.next
        if t == self.head:
          return -1
      t.prev.next = t.next
      t.next.prev = t.prev
      t.next = None
      t.prev = None


circular_doubly_linked_list  = cdlinkedlist()

circular_doubly_linked_list.addfirst(5)
circular_doubly_linked_list.addfirst(3)
circular_doubly_linked_list.addfirst(8)
circular_doubly_linked_list.addfirst(2)
circular_doubly_linked_list.addfirst(1)

circular_doubly_linked_list.display()

circular_doubly_linked_list.delete(3)

circular_doubly_linked_list.display()
```
